###Cantidad de episodios por año

```python
#Pasamos la columna fecha a tipo datetime
himym_episodes["original_air_date"]= pd.to_datetime(himym_episodes["original_air_date"])
```

```python
himym_episodes['year'] = himym_episodes["original_air_date"].dt.strftime('%Y')
himym_episodes['month'] = himym_episodes["original_air_date"].dt.strftime('%m')
```

```python
#Cantidad de episodios por año
episode_by_year = himym_episodes.groupby(by="year").size().reset_index(name="Cantidad de episodios")
episode_by_year
```

###Cantidad de espectadores por temporada

```python
episode_by_viewers = himym_episodes.groupby(by="season").sum()
episode_by_viewers.us_viewers
```

###Join de dataframe

```python
#Incorporamos índice para hacer el merge de los dos dataframe
#himym_imdb['index'] = himym_imdb.index
#himym_imdb['index']  = np.arange(1, len(himym_imdb)+1)
#himym_imdb
```

```python
#join by two columns
imdb_episodes  = pd.merge(himym_imdb, himym_episodes, left_on=["season","episode_num"], right_on=["season","episode_num_in_season"])
imdb_episodes.info()
```

###Episodios con más ranking

```python
imdb_episodes_mask = imdb_episodes["imdb_rating"] == imdb_episodes["imdb_rating"].max()
positions = np.flatnonzero(imdb_episodes_mask)
filtered_imdb_episodes = imdb_episodes.iloc[positions]
filtered_imdb_episodes
```

###Episodio con menos ranking

```python
imdb_episodes_mask_2 = imdb_episodes["imdb_rating"] == imdb_episodes["imdb_rating"].min()
positions_2 = np.flatnonzero(imdb_episodes_mask_2)
filtered_imdb_episodes_2 = imdb_episodes.iloc[positions_2]
filtered_imdb_episodes_2
```

###Episodio con más viewers

```python
imdb_episodes_mask_3 = imdb_episodes["us_viewers"] == imdb_episodes["us_viewers"].max()
positions_3 = np.flatnonzero(imdb_episodes_mask_3)
filtered_imdb_episodes_3= imdb_episodes.iloc[positions_3]
filtered_imdb_episodes_3
```

###Episodio con menos viewers

```python
imdb_episodes_mask_4 = imdb_episodes["us_viewers"] == imdb_episodes["us_viewers"].min()
positions_4 = np.flatnonzero(imdb_episodes_mask_4)
filtered_imdb_episodes_4= imdb_episodes.iloc[positions_4]
filtered_imdb_episodes_4
```

###Episodio con más votos

```python
imdb_episodes_mask_5 = imdb_episodes["total_votes"] == imdb_episodes["total_votes"].max()
positions_5 = np.flatnonzero(imdb_episodes_mask_5)
filtered_imdb_episodes_5= imdb_episodes.iloc[positions_5]
filtered_imdb_episodes_5
```

###Episodio con menos votos

```python
imdb_episodes_mask_6 = imdb_episodes["total_votes"] == imdb_episodes["total_votes"].min()
positions_6 = np.flatnonzero(imdb_episodes_mask_6)
filtered_imdb_episodes_6= imdb_episodes.iloc[positions_6]
filtered_imdb_episodes_6
```

##Visualizaciones

###N° Episodios por Director

```python
chart_1=alt.Chart(imdb_episodes).mark_bar().encode(
    x='directed_by',
    y='episode_num',
    color='directed_by'
).properties(title= "N° Episodios por Director", height=300, width=400)

chart_1.configure_title(
    fontSize=20,
    font='Arial',
    color='gray',
    align = 'center'
)

chart_1.configure_axis(
    labelAngle=0)
```

###N° Espectadores por Temporada

```python
alt.Chart(imdb_episodes).mark_bar().encode(
    x='season',
    y='us_viewers'
).properties(title= "N° de Espectadores por Temporada", height=300, width=400)
```

```python
alt.Chart(imdb_episodes).mark_bar().configure_axis(
    labelAngle=0
).encode(
    x='directed_by',
    y='us_viewers',
    color='directed_by'
).properties(title= "N° de Espectadores por Director", height=300, width=400)
```

###N° de Espectadores por Año

```python
alt.Chart(imdb_episodes).mark_bar().configure_axis(
    labelAngle=0
    ).encode(
    x='year',
    y='us_viewers'
).properties(title= "N° de Espectadores por Año", height=300, width=400)
```
